```python
import pandas as pd
import random
```

```python
def air_lines():
    d={}
    while True:
        try:
            key=input('enter flightname(to stop-z):')
        except ValueError:
            print('enter valid name')
        if key.lower()=='z':
            break
        val={}
        while True:
            k=input('enter class(to stop-y):')
            if k=='y':
                break
            try:
              v=int(input(f'enter no of seats available in {k}:'))
            except ValueError:
              print('enter valid seat')
            val[k]=v
        d[key]=val
    return d

```

```python
air_=air_lines()
df=pd.DataFrame(air_)
display(df)
```

```python
df2=pd.DataFrame(columns=['Name','Age','Gender','FlightName','Ticket'])
```

```python
def flights_available():
    count=1
    for airline in df.columns:
        print(f'{count}.',airline)
        count+=1
```

```python
def class__():
    count=1
    for classes in df.index:
        print(f'{count}.',classes)
        count+=1
```

```python
unique=set()
valid_chars = 'abcdefghijklmnopqrstuvwxyz'
num_='0123456789'
def gener_ticket(class_name):
    p1=''.join(random.choices(valid_chars, k=2))
    p2=''.join(random.choices(num_, k=3))
    pwd=p1+p2
    if class_name=='First':
        ticket='First-'f'{pwd}'
    elif class_name=='Business':
        ticket='Business-'f'{pwd}'
    elif class_name=='Economy':
        ticket='Economy-'f'{pwd}'
    if ticket not in unique:
        unique.add(ticket)
        return ticket
    else:
        return gener_ticket(class_name)
```

```python
def customer_detail():
    d={}
    key=0
    while True:
        print(df)
        key=key+1
        name1=input('enter name (to end press-z):')
        if name1.lower()=='z':
            break
        namecheck=input('Please check your name.Press(y-to correct it / n-to continue):')
        if namecheck.lower()!='n':
            name1=input('enter name (to end press-z):')
        age=int(input('enter age (only in numbers) :'))
        gender=input('enter gender(F/M/Others):')
        flights_available()
        choice=input('enter flight name (the same as mentioned in options):')
        fltcheck=input('Please check the entered flight name.Press(y-to correct it / n-to continue):')
        if fltcheck.lower()!='n':
            flights_available()
            choice=input('enter flight name (the same as mentioned in options):')
        class__()
        class_=input('enter class (the same as mentioned in the options):')
        classcheck=input('Please check the entered class name.Press(y-to correct it / n-to continue):')
        if classcheck.lower()!='n':
            class__()
            class_=input('enter class (the same as mentioned in options):')
        if df[choice][class_]==0:
            print(f'{class_} not available under {choice}')
            print('choose another class')
            class__()
            class_=input(f'enter class other than {class_}:')
        ticket_=gener_ticket(class_)
        print(f'Generated ticket: {ticket_}')
        print('Your Booking is confirmed')
        d[key]={'Name':name1,'Age':age,'Gender':gender,'FlightName':choice,'Ticket':ticket_}
        df[choice][class_]-=1
    return d,df
```
